```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d = "/home/abhirup.ghosh/Documents/Work/spinqnm/runs/parspec/injections/GW150914/zeronoise_highSNR/cbcBayes/posterior_samples.dat"
d = np.genfromtxt(d, dtype=None, names=True)
```

```python
param_list = ['alphangr', 'mf_evol', 'af_evol', 'mtotal', 'm1', 'm2', 'q', 'redshift', 'dist', 'chi_eff']
```

```python
%matplotlib inline

idx, = np.where(d['alphangr'] < 10)
idx1, = np.where(d['alphangr'] > 10)

fig, axs = plt.subplots(ncols=len(param_list), nrows=1, figsize=(40,4))

for i, par in enumerate(param_list):
    
    sns.histplot(d[par][idx], ax=axs[i], stat='probability', color="#102542", element='step', alpha=0.5)
    sns.histplot(d[par][idx1], ax=axs[i], stat='probability', color="#f87060", element='step', alpha=0.5)
    
    axs[i].set_xlabel(par)

plt.tight_layout()
```
