```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('../data/O3list.dat', names=True, dtype=None)
```

```python
data['Event']
```

```python
plt.figure(figsize=(15,10))
for i in range(len(data)):
    x = data['M'][i]
    y = data['IFAR'][i]
    plt.semilogy(x, y, marker='.', markersize=30,color='red')
    plt.text(x+0.3, y+0.3, data['Event'][i], fontsize=9)
plt.axvline(x=50)    
plt.xlabel('Total Mass (GraceDB template) [Msun]')
plt.ylabel('IFAR')    
plt.savefig('./O3a_tgr_event_list_M_IFAR.png', dpi=130)
```
