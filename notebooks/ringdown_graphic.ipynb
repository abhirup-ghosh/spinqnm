```python
import numpy as np
import lalsimulation
import lal
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Times New Roman"
```

```python
def qnm(t, f, tau):
    return np.exp(-t/tau)*np.cos(2*np.pi*f*t)
```

```python
%matplotlib inline

t = np.linspace(0,0.1,1000)

fig, ax = plt.subplots(figsize=(10,8))

for (f, tau) in zip(np.linspace(230,290,50),np.linspace(0.005,0.01,50)):
    h = qnm(t,f,tau)
    ax.plot(t,h,color='k',alpha=0.3,lw=0.5)
    
ax.set_xlim(0, 0.06)
#ax.text(0.03,0.9,'Probing the nature of gravity \nusing the...',ha='center',va='center',fontsize=25,style='italic')
#ax.text(0.05,0.06,'...black-hole',ha='center',va='center',fontsize=25)
#ax.text(0.05,-0.06,'RINGDOWN.',ha='center',va='center',fontsize=25,style='italic')
#ax.text(0.03,-0.7,'Was Einstein wrong? \nAnd someone else right?',ha='center',va='center',fontsize=25,style='italic')
#ax.text(0.058,-0.8,'Credit: Abhirup Ghosh/AEI-Potsdam',fontstyle='italic',ha='right',color='k',fontsize=10,rotation=90)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.set_xticks([])
ax.set_yticks([])

plt.tight_layout()
plt.savefig("/Users/abhirupghosh/Desktop/ringdown.png",dpi=400)
```
