```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt("../data/benchmarking.dat", dtype=None, names=True)
data['time']-data['time'][0]
```

```python
plt.figure(figsize=(5,5))
plt.semilogy((data['time']-data['time'][0])/ 60. / 60., data['lc1'], marker='o')
plt.semilogy((data['time']-data['time'][0])/ 60./ 60., data['lc2'], marker='o')
plt.semilogy((data['time']-data['time'][0])/ 60./ 60., data['lc3'], marker='o')
plt.xlabel("time (hours)")
plt.ylabel("samples accumulated")
plt.show()
```
