```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import lal
```

```python
data = np.genfromtxt("posterior_samples_p4_df0_dtau0_l.dat", names=True, dtype=None)
```

```python
df, dtau, l, m2s = data['domega220'], data['dtau220'], data['alphangr'], data['m2_source']

# converting to SI units
m2s = m2s*1.47
m2s_med = np.median(m2s)
```

```python
p = 4
```

```python
fig, ax = plt.subplots(figsize=(10,8))

sns.histplot(l, kde="True", element="step", ax=ax)
ax.axvline(x=(0.5*m2s_med**p)**(1/p), color='k', ls='--', label="$[0.5 \, m_{s}^{p}]^{1/p}$")
#ax.axvline(x=(0.5*(29.*1.47)**p)**(1/p), color='r', ls='--', label="$[0.5 \, m_{s}^{p}]^{1/p}$")
ax.set_xlabel('$\ell$', fontsize=20)
ax.set_ylabel('PDF', fontsize=20)
ax.legend(loc='best', fontsize=20)

#ax.set_xlim([min(l), max(l)])

for spine in plt.gca().spines.values():
    spine.set_visible(False)

plt.tight_layout()    
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,8))

sns.histplot((l/m2s)**p, kde="True", element="step", ax=ax)
ax.axvline(x=0.5, color='k', ls='--')
ax.set_xlabel('$\ell$', fontsize=20)
ax.set_ylabel('PDF', fontsize=20)
ax.legend(loc='best', fontsize=20)

#ax.set_xlim([min(l), max(l)])

for spine in plt.gca().spines.values():
    spine.set_visible(False)

plt.tight_layout()    
plt.show()
```
