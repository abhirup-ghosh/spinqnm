```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# import lal
```

```python
# data = np.genfromtxt("posterior_samples_p4_df0_dtau0_l.dat", names=True, dtype=None)
data = np.genfromtxt("theory_specific_samples/eft_of_gr_cubic_even/posterior_samples.dat", names=True, dtype=None)
```

```python
df, dtau, l, m2s, M, z = data['domega220'], data['dtau220'], data['alphangr'], data['m2_source'], data['mtotal'], data['redshift']

# converting to SI units
m2s = m2s*1.47
m2s_med = np.median(m2s)
```

```python
p = 4
```

```python
fig, ax = plt.subplots(figsize=(10,8))

sns.histplot(l, kde="True", element="step", ax=ax)
ax.axvline(x=(0.5*m2s_med**p)**(1/p), color='k', ls='--', label="$[0.5 \, m_{s}^{p}]^{1/p}$")
ax.set_xlabel('$\ell$', fontsize=20)
ax.set_ylabel('PDF', fontsize=20)
ax.legend(loc='best', fontsize=20)

ax.set_xlim([min(l), max(l)])

ax.set_xlabel(r'$\lambda_{\rm e}^{1/4} \, \ell$ [km]')

# for spine in plt.gca().spines.values():
#     spine.set_visible(False)

plt.tight_layout()    
plt.show()

plt.savefig('cubic_eft_in_progress.png')
```

```python
fig, ax = plt.subplots(figsize=(10,8))

sns.histplot((l/m2s)**p, kde="True", element="step", ax=ax)
ax.axvline(x=0.5, color='k', ls='--')
ax.set_xlabel(r'$\lambda_{\rm e}^{1/4} \, \ell/m^s_2$', fontsize=20)
ax.set_ylabel('PDF', fontsize=20)
ax.legend(loc='best', fontsize=20)

#ax.set_xlim([min(l), max(l)])

for spine in plt.gca().spines.values():
    spine.set_visible(False)

plt.tight_layout()    
plt.show()
```
