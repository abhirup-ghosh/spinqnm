```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_0p0 = '../runs/nonGR/test/0p0'
data_0p5 = '../runs/nonGR/test/0p5'
```

```python
color = ['k', 'r']
ls = ['solid', 'dashed']
label = ['no deviations', 'domega = dtau = 0.5']

trigtime = 1126258000

for ifo in ['H1','L1','V1']:
    plt.figure(figsize=(5,5))
    for (idx, data) in enumerate([data_0p0, data_0p5]):
        t, hf = np.loadtxt(data+'/%s_timeInjection.dat'%ifo, unpack=True)
        plt.plot(t, hf, color=color[idx], ls=ls[idx], label=label[idx])
        
        plt.xlim([trigtime-0.1,trigtime+0.01])
        plt.title('%s'%ifo)
    plt.show()
```
