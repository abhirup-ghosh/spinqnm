```python
import numpy as np
import matplotlib.pyplot as plt
```

# GW150914-like

```python
loc_gr = '../runs/nonGR/GW150914-like/0p0'
loc_ngr = '../runs/nonGR/GW150914-like/0p5'

label_list = ['GR','nGR']
ls_list = ['solid', 'dashed']
```

```python
%matplotlib notebook
%matplotlib notebook

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)


for (idx,loc) in enumerate([loc_gr, loc_ngr]):
        t, ht = np.genfromtxt(loc + '/H1_timeInjection.dat', unpack=True)
        
        ax1.plot(t,ht, ls=ls_list[idx], label=label_list[idx])
        ax1.set_title('H1')

        t, ht = np.genfromtxt(loc + '/L1_timeInjection.dat', unpack=True)
        ax2.plot(t,ht, ls=ls_list[idx], label=label_list[idx])
        ax2.set_title('L1')

plt.legend(loc='best')    
```
